```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pickle as pk
from pytube import YouTube
```

```python
data_root='/workspace/hdd1/230406_joseph_mel/acav100m/data/'
save_root='/workspace/hdd1/230406_joseph_mel/acav100m/data/'

if not os.path.isdir(os.path.join(save_root,'video200k')):
    os.makedirs(os.path.join(save_root,'video200k'))

annot_csvs = os.path.join(data_root, 'ncentroids-500-subset_size-200K.csv')
link_form = 'https://www.youtube.com/watch?v='
saved_root=os.listdir(os.path.join(save_root,'video200k'))
count=0
with open(annot_csvs) as f:
    lines = f.read().splitlines()
    total = len(lines)
    curr_idx=0
    for row in lines:
        curr_idx+=1
        print('{}/{}'.format(curr_idx,total))
        fname = row.split(',')[0]
        time = row.split(',')[1]
        curr_link = link_form+fname
        print(fname+'.mp4')
        ffname=fname+'.mp4'
        if ffname in saved_root:
            print("on")
            continue
        else:    
            try:
                print("yes")
                yt = YouTube(curr_link,use_oauth=True, allow_oauth_cache=True)
                yt.streams.get_highest_resolution().download(output_path=os.path.join(save_root,'video200k'),filename=fname+".mp4")
            except:
                continue
print(count)
```
